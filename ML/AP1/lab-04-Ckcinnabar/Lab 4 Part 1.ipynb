# Problem 1 (10 points)

**Recall that the Support Vector Machine objective function is**

$$C\sum_{n=1}^N \xi_n + \frac{1}{2} \Vert w\Vert^2$$

**subject to the constraints**

\begin{align}
t_n y(x_n) &\geq 1 - \xi_n, n = 1, \dots, N \\
\xi_n &\geq 0, n = 1, \dots, N
\end{align}

**Answer the following questions:**
1. **Describe the role and purpose of each term and each constraint in the SVM objective function?**



$$\frac{1}{2} \Vert w\Vert^2$$ 

This regularization term controls the margin width. Since margin width is inversely proportional to ∥w∥, minimizing ∥w∥^2 maximizes the margin. A larger margin generally leads to better generalization.

$$C\sum_{n=1}^N \xi_n$$


This penalty term controls the cost of misclassification. The slack variables $$ \xi_n$$ ​ allow data points to violate the margin or be misclassified. 

$$t_n y(x_n)$$


This ensures that each point is either correctly classified beyond the margin or incurs a penalty via the slack variable​. The constraint links the classification result to the error penalty.

$$ \xi_n$$
 This ensures slack variables are non-negative, as they represent the degree of margin violation.

2. **$C$ is a parameter that is set by the user. Describe the relationship between values of $C$ and the resulting SVM decision surface, performance and number of support vectors.**

Large C values: Typically results in fewer support vectors, as the model tries to correctly classify most points beyond the margin


Small C values: Typically results in more support vectors, as more points are allowed to be on or within the margin

# Problem 2 (10 points)

**Suppose you have a binary (2-class) classification problem. Answer the following questions**

1. **Describe how the logistic regression and the SVM find a decision surface to classify the two classes.**


Logistic Regression finds a decision boundary by modeling the probability of class membership through the sigmoid function, setting the boundary where the probability equals 0.5. 

SVM takes a geometric approach by finding the hyperplane that maximizes the margin between classes, focusing on the support vectors closest to the boundary and can employ kernel functions to create non-linear boundaries.

2. **When would you choose SVM over Logistic Regression and vice-versa? Justify your answers.**

SVM generally performs better when working with high-dimensional data, when a non-linear decision boundary is needed, or when robustness against outliers is important. 

Logistic Regression is typically preferred when you need probability estimates rather than just classifications, when model interpretability is important, or when dealing with large datasets that require frequent updates.

## Problem 4 (10 points)
Answer the following questions:
1. What are the common challenges when training decision tree classifier?

Decision trees suffer from overfitting, instability (small data changes causing different trees), bias toward features with more splits, and creating only orthogonal decision boundaries.


2. Discuss how a random forest classifier help address these challenges.

Random forests combine multiple trees trained on bootstrap samples with random feature selection at each split, reducing overfitting and instability while improving accuracy through ensemble averaging.

3. Are random forests interpretable?

Random forests offer moderate interpretability through feature importance measures and partial dependence plots, though not as transparent as single trees but more interpretable than black-box models like neural networks.

## Problem 5 (10 points)
Answer the following questions
1. What is ensemble learning, and how does it improve the performance of a model?

Ensemble learning combines multiple models to produce better predictions than any single model. It improves performance by reducing variance through averaging, decreasing bias through diverse models, and canceling out individual model errors.


2. Explain the difference between bagging and boosting.

Bagging trains models independently in parallel on random data subsets and combines them through averaging. Boosting trains models sequentially, with each new model focusing on correcting previous models' errors by emphasizing misclassified instances.

3. How does pasting differ from bagging?

Pasting samples training data without replacement (each instance appears at most once), while bagging uses sampling with replacement (bootstrapping), allowing duplicate instances and creating more diverse training subsets.

4. What are the advantages of using ensemble methods compared to a single model?

Ensemble methods provide higher accuracy, better generalization, reduced overfitting, greater robustness to noise, and more stable results across different datasets. They effectively combine strengths of different models while minimizing individual weaknesses.

5. In which situations would bagging be preferable over boosting, and vice versa?

Bagging works better with high-variance models and noisy data with outliers. Boosting excels with weak, high-bias models and clean data when maximum accuracy is needed, though it risks overfitting.
