## Session 2. Getting Started with Libraries and Frameworks

This Jupyter Notebook serves as an introductory playground to familiarize you with some of the key libraries we will be using throughout the semester in our machine learning projects. Additionally, it provides an overview of the end-to-end machine learning process.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

```

```python
# Step 1: Load the Iris dataset using scikit-learn
iris = load_iris()  # Load the Iris dataset into memory
X, y = iris.data, iris.target  # Separate the features (X) and the target labels (y)

X
```

```python
y
```

```python
# Step 2: Convert to pandas DataFrame for easier manipulation
df = pd.DataFrame(X, columns=iris.feature_names)  # Convert features into a DataFrame with column names
df['target'] = y  # Add the target labels as a new column in the DataFrame
```

```python
df
```

```python
# Step 3: Detect and remove outliers using SciPy (Z-score method)
z_scores = np.abs(stats.zscore(df.iloc[:, :-1]))  # Calculate Z-scores for feature columns
df_no_outliers = df[(z_scores < 3).all(axis=1)]  # Remove rows where any feature's Z-score exceeds 3
```

```python
df_no_outliers
```

```python
# Step 4: Split data into features and target after outlier removal
X_clean = df_no_outliers.iloc[:, :-1].values  # Extract cleaned feature values
y_clean = df_no_outliers['target'].values  # Extract the cleaned target labels
```

```python
X_clean
```

```python
y_clean
```

```python
# Step 5: Train-test split using scikit-learn
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)  # Split data
```

```python
X_train
```

```python
X_train.shape
```

```python
# Step 6: Standardize features using scikit-learn's StandardScaler
scaler = StandardScaler()  # Initialize the scaler
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform training data
X_test_scaled = scaler.transform(X_test)  # Transform testing data
```

```python
X_train_scaled 
```

```python
# Step 8: Train Logistic Regression model using scikit-learn
model = LogisticRegression()  # Initialize Logistic Regression model
model.fit(X_train_scaled, y_train)  # Train the model on the PCA-transformed data
y_pred = model.predict(X_test_scaled)  # Predict the labels for the test set
```

```python
# Step 9: Generate classification report using scikit-learn
print("Classification Report:\n")  # Print header
print(classification_report(y_test, y_pred))  # Display precision, recall, and f1-score
```

```python
# Step 10: Create a confusion matrix heatmap for model performance using Matplotlib
conf_matrix = confusion_matrix(y_test, y_pred)  # Generate confusion matrix
plt.figure(figsize=(6, 5))  # Set figure size
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)  # Plot confusion matrix as heatmap
plt.title('Confusion Matrix Heatmap')  # Set title
plt.colorbar()  # Add color bar to indicate intensity
classes = iris.target_names
plt.xticks(np.arange(len(classes)), classes, rotation=45)  # Set x-ticks
plt.yticks(np.arange(len(classes)), classes)  # Set y-ticks

# Annotate the heatmap with counts
thresh = conf_matrix.max() / 2.
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(j, i, format(conf_matrix[i, j], 'd'),
                 ha='center', va='center',
                 color='white' if conf_matrix[i, j] > thresh else 'black')

plt.ylabel('True Labels')  # Label y-axis
plt.xlabel('Predicted Labels')  # Label x-axis
plt.tight_layout()  # Adjust layout
plt.show()  # Display the plot
```
