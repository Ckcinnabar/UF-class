**Answer the following questions:**

1. **Which algorithm(s), if any, identifies the intrinsic manifold structure of the data in a 2-D projection. Explain your answer.**

ISOMAP best identifies the intrinsic manifold structure because it preserves the circular shape with a hollow center and maintains proper point relationships by using geodesic distances along the manifold surface.

2. **What assumptions does MDS (with Euclidean distance), ISOMAP and LLE make about the data embedding space?**

MDS with Euclidean distance assumes data lies in a linear space where straight-line distances reflect true similarities, making it struggle with curved manifolds.


ISOMAP assumes data lies on a manifold that is locally Euclidean but globally curved, calculating distances along the manifold surface rather than through space.


LLE assumes data points and their neighbors lie on or near a locally linear patch of a manifold, focusing exclusively on preserving local neighborhood relationships without making global distance assumptions

3. **How is the local and global data structure being preserved for each algorithm?**

MDS (Euclidean distance):

Global structure: Attempts to preserve all pairwise Euclidean distances between points, maintaining global relationships

Local structure: Often sacrifices local relationships to minimize global distance distortion

ISOMAP:

Global structure: Preserves geodesic distances (shortest paths along the manifold) between all pairs of points

Local structure: Constructs local neighborhoods first, then uses them to estimate global geodesic distances

LLE:

Global structure: Does not explicitly preserve global structure or distances

Local structure: Strongly preserves local neighborhood relationships by representing each point as a weighted combination of its neighbors

4. **Discuss advantages and disadvantages for each algorithm.**

MDS (Euclidean Distance)
Advantages:

Conceptually simple and easy to implement, works well for data with linear structure

Disadvantages:

Struggles with non-linear manifolds, computationally expensive for large datasets

ISOMAP
Advantages:

Effectively preserves both local and global structure of non-linear manifolds, uses geodesic distances which better represent true relationships on curved surfaces

Disadvantages:

Sensitive to the choice of neighborhood size, computationally expensive (requires shortest path calculations)

LLE
Advantages:

Excellent at preserving local neighborhood relationships, computationally more efficient than ISOMAP

Disadvantages:

Global structure can be significantly distorted, very sensitive to the choice of number of neighbors
