##Course Policies
###1. Are late assignments acceptable?
Nope
###2. What to do if you feel that an assignment needs to be re-graded?
We should contact the professor directly to discuss any grading concerns
###3. What constitutes healthy collaboration and when to cite another student?
Talking about the problem
Using a whiteboard (or paper, or something similar) to draw out the problem
Looking at someone else’s work to help them identify or fix a bug, AFTER you have already completed that portion for yourself
