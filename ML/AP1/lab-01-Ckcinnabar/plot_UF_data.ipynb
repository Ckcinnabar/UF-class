# UF_data Visualization

Scatter plot of `UF_data.npy` (shape `(3000, 3)`): X, Y coordinates colored by class label.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load("UF_data.npy")
data.shape
```

```python
x = data[:, 0]
y = data[:, 1]
label = data[:, 2]
```

```python
# UF official colors
UF_BLUE = "#0021A5"
UF_ORANGE = "#FA4616"

plt.figure(figsize=(10, 8))
for value, color, name in [(0, UF_BLUE, "Class 0"), (1, UF_ORANGE, "Class 1")]:
    mask = label == value
    plt.scatter(x[mask], y[mask], c=color, s=10, label=name)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("UF Dataset Scatter Plot")
plt.legend()
plt.show()
```
