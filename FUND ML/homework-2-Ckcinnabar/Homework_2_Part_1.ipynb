# Exercise 4 (18 points)

**For each statement below, state whether it is true (T) or false (F) and provide a justification for your answer.**

1. (2 points) **Minimizing the Least Squares objective function with a weight decay term (regularizer) is equivalent to maximizing the data likelihood with a prior on the weights.**

**Answer: True (T)**

**Justification:**

Minimizing the Least Squares objective with L2 regularization (weight decay):
$$\min_{\mathbf{w}} \sum_{i=1}^N (y_i - \mathbf{w}^T\mathbf{x}_i)^2 + \lambda ||\mathbf{w}||^2$$

is equivalent to maximizing the posterior in a Bayesian framework:
$$\max_{\mathbf{w}} P(\mathbf{w}|\mathbf{X}, \mathbf{y}) = \max_{\mathbf{w}} P(\mathbf{y}|\mathbf{X}, \mathbf{w})P(\mathbf{w})$$

where:
- The likelihood $P(\mathbf{y}|\mathbf{X}, \mathbf{w}) \propto \exp\left(-\frac{1}{2\sigma^2}\sum_i (y_i - \mathbf{w}^T\mathbf{x}_i)^2\right)$ (Gaussian noise assumption)
- The prior $P(\mathbf{w}) \propto \exp\left(-\frac{\lambda}{2}||\mathbf{w}||^2\right)$ (Gaussian prior on weights)

Taking the negative log of the posterior gives the regularized least squares objective. This is the MAP (Maximum A Posteriori) estimation.

2. (2 points) **The observed data likelihood on $N$ i.i.d. samples $x_i$ (where $x_i\geq0, \forall i$) with a Gamma distribution, $P(x|\alpha,\beta)=\frac{\beta^{\alpha}}{\Gamma(\alpha)}x^{\alpha-1}e^{-\beta x}$, and a prior probability on parameter $\beta$ as another Gamma distribution, $P(\beta|a,b)=\frac{b^a}{\Gamma(a)}\beta^{a-1}e^{-b \beta}$, form a Conjugate Prior relationship.**

**Answer: True (T)**

**Justification:**

For a conjugate prior relationship, the posterior must have the same functional form as the prior.

Given:
- Likelihood: $P(\mathbf{x}|\alpha,\beta) = \prod_{i=1}^N \frac{\beta^{\alpha}}{\Gamma(\alpha)}x_i^{\alpha-1}e^{-\beta x_i}$
- Prior: $P(\beta|a,b) = \frac{b^a}{\Gamma(a)}\beta^{a-1}e^{-b\beta}$

The posterior is:
$$P(\beta|\mathbf{x},\alpha,a,b) \propto P(\mathbf{x}|\alpha,\beta)P(\beta|a,b)$$

$$\propto \beta^{N\alpha} e^{-\beta\sum_i x_i} \cdot \beta^{a-1}e^{-b\beta}$$

$$= \beta^{(N\alpha + a - 1)}e^{-\beta(\sum_i x_i + b)}$$

This is a Gamma distribution with parameters $(a' = N\alpha + a, b' = \sum_i x_i + b)$, which has the same form as the prior. Therefore, the Gamma distribution is a conjugate prior for the parameter $\beta$ in the Gamma likelihood (when $\alpha$ is known).

3. (2 points) **In MLE estimation we are optimizing the posterior distribution.**

**Answer: False (F)**

**Justification:**

In Maximum Likelihood Estimation (MLE), we optimize the **likelihood function** $P(\mathbf{X}|\theta)$, not the posterior distribution.

$$\theta_{\text{MLE}} = \arg\max_{\theta} P(\mathbf{X}|\theta)$$

The **posterior distribution** $P(\theta|\mathbf{X})$ is optimized in Maximum A Posteriori (MAP) estimation:

$$\theta_{\text{MAP}} = \arg\max_{\theta} P(\theta|\mathbf{X}) = \arg\max_{\theta} P(\mathbf{X}|\theta)P(\theta)$$

MLE does not consider a prior distribution $P(\theta)$, while MAP does. MLE is a frequentist approach, whereas MAP is a Bayesian approach.

4. (2 points) **Consider $\theta_{\text{MLE}} = \arg_{\theta} \max P(X|\theta)$ and $\theta_{\text{MAP}} = \arg_{\theta} \max P(X|\theta)P(\theta)$. We have that $\theta_{\text{MLE}} = \theta_{\text{MAP}}$ for uniform-distributed priors.**

**Answer: True (T)**

**Justification:**

For a uniform prior $P(\theta) = c$ (constant), the MAP estimation becomes:

$$\theta_{\text{MAP}} = \arg\max_{\theta} P(X|\theta)P(\theta) = \arg\max_{\theta} P(X|\theta) \cdot c$$

Since $c$ is a constant and does not depend on $\theta$, it doesn't affect the location of the maximum:

$$\theta_{\text{MAP}} = \arg\max_{\theta} P(X|\theta) = \theta_{\text{MLE}}$$

Therefore, when the prior is uniform (uninformative), MAP estimation reduces to MLE estimation. The uniform prior represents no preference for any particular parameter value, making MAP equivalent to MLE.

5. (2 points) **The EM algorithm guarantees convergence to the global optima solution.**

**Answer: False (F)**

**Justification:**

The EM algorithm is guaranteed to converge to a **local optimum**, not necessarily the global optimum.

Key properties of EM:
- Each iteration of EM is guaranteed to increase (or maintain) the log-likelihood: $\mathcal{L}(\Theta^{t+1}) \geq \mathcal{L}(\Theta^t)$
- EM will converge to a stationary point of the likelihood function
- The final solution depends on the initialization of parameters

Since the likelihood function for mixture models is typically non-convex with multiple local maxima, EM can get stuck in local optima. Different initializations can lead to different solutions. This is why:
1. Multiple random initializations are often used
2. K-means initialization is common for GMMs
3. The best result among multiple runs is typically selected

6. (2 points) **The latent variables $z$ in the EM algorithm are always discrete random variables.**

**Answer: False (F)**

**Justification:**

Latent variables $z$ in the EM algorithm can be either **discrete or continuous** random variables.

**Discrete latent variables examples:**
- **Gaussian Mixture Models (GMM)**: $z_i \in \{1, 2, ..., K\}$ indicates which component generated data point $x_i$
- **Hidden Markov Models**: discrete hidden states

**Continuous latent variables examples:**
- **Factor Analysis**: continuous latent factors
- **Probabilistic PCA**: continuous latent representations
- **Variational Autoencoders**: continuous latent codes in the latent space

The EM algorithm framework applies to both cases. The main difference is:
- For discrete $z$: E-step computes $P(z|x, \theta^t)$ as a probability distribution over discrete states
- For continuous $z$: E-step computes expectations with respect to a continuous distribution $P(z|x, \theta^t)$

7. (2 points) **We can use the EM algorithm to optimize a Gamma mixture model, $P(x) = \sum_{k=1}^K \pi_k \Gamma\left(x|\nu_k,\theta_k\right)$ to a set of $N$ i.i.d. samples $\{x_i\}_{i=1}^N$.**

**Answer: True (T)**

**Justification:**

Yes, the EM algorithm can be used to optimize a Gamma mixture model. The framework of EM applies to any mixture model where:

1. **Observed data likelihood** can be written as a mixture:
   $$P(x) = \sum_{k=1}^K \pi_k \Gamma(x|\nu_k, \theta_k)$$

2. **Complete data likelihood** includes latent indicators $z_i \in \{1,...,K\}$:
   $$P(x_i, z_i) = \pi_{z_i} \Gamma(x_i|\nu_{z_i}, \theta_{z_i})$$

3. **E-step** computes the responsibility (membership probability):
   $$C_{ik} = P(z_i=k|x_i, \Theta^{(t)}) = \frac{\pi_k \Gamma(x_i|\nu_k, \theta_k)}{\sum_{j=1}^K \pi_j \Gamma(x_i|\nu_j, \theta_j)}$$

4. **M-step** updates parameters by maximizing:
   $$Q(\Theta, \Theta^{(t)}) = \sum_{i=1}^N \sum_{k=1}^K C_{ik} \log[\pi_k \Gamma(x_i|\nu_k, \theta_k)]$$

The EM algorithm works for any parametric mixture model with well-defined likelihood functions.

8. (2 points) **The MLE solution for the parameter $\theta_k$ in the Gamma mixture model is $\theta_k = \frac{\nu_k \sum_{i=1}^N C_{ik}}{\sum_{i=1}^N x_i C_{ik}}$.**

**Answer: True (T)**

**Justification:**

For a Gamma distribution $\Gamma(x|\nu, \theta) = \frac{\theta^{\nu}}{\Gamma(\nu)}x^{\nu-1}e^{-\theta x}$, we can verify this is the correct MLE update for $\theta_k$ in the M-step.

The Q-function for the Gamma mixture model includes terms:
$$Q \propto \sum_{i=1}^N \sum_{k=1}^K C_{ik} \log\left[\frac{\theta_k^{\nu_k}}{\Gamma(\nu_k)}x_i^{\nu_k-1}e^{-\theta_k x_i}\right]$$

$$= \sum_{i=1}^N \sum_{k=1}^K C_{ik} \left[\nu_k \log \theta_k - \theta_k x_i + \text{const}\right]$$

Taking the derivative with respect to $\theta_k$ and setting to zero:
$$\frac{\partial Q}{\partial \theta_k} = \sum_{i=1}^N C_{ik}\left[\frac{\nu_k}{\theta_k} - x_i\right] = 0$$

$$\frac{\nu_k}{\theta_k}\sum_{i=1}^N C_{ik} = \sum_{i=1}^N C_{ik} x_i$$

$$\theta_k = \frac{\nu_k \sum_{i=1}^N C_{ik}}{\sum_{i=1}^N x_i C_{ik}}$$

This matches the given formula exactly.

9. (2 points) **A Gaussian Mixture Model can be used to find cluster groups in the data.**

**Answer: True (T)**

**Justification:**

A Gaussian Mixture Model (GMM) is indeed commonly used for clustering. Here's why:

**GMM as a Clustering Method:**
1. **Soft Clustering**: GMM provides probabilistic cluster assignments through $C_{ik} = P(z_i=k|x_i)$, which represents the probability that data point $x_i$ belongs to cluster $k$

2. **Hard Clustering**: For hard cluster assignment, we can assign each point to the cluster with highest probability:
   $$\text{cluster}(x_i) = \arg\max_k C_{ik}$$

3. **Advantages over K-means**:
   - Captures elliptical cluster shapes (not just spherical)
   - Provides uncertainty estimates through probabilities
   - Models clusters with different sizes and covariances

4. **Interpretation**: Each Gaussian component represents a cluster, with:
   - $\mu_k$: cluster center
   - $\Sigma_k$: cluster shape and spread
   - $\pi_k$: cluster proportion/weight

GMM is one of the most popular model-based clustering algorithms in machine learning and statistics.
