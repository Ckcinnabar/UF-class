## Exercise 2 (3 points)

**Consider the data matrix $\mathbf{X}$ of size $3\times N$, where $N$ is the number of samples. The covariance matrix $\mathbf{K}$, of size $3\times 3$ has 3 eigenvectors $v_1=[-0.99, 0.09, 0]^T$, $v_2=[0,0,1]^T$ and $v_3=[-0.09, -0.99, 0]^T$ with eigenvalues $\lambda_1=0.98$, $\lambda_2=0.5$ and $\lambda_3=1.98$, respectively. Answer the following questions:**

1. (1 point) **What linear transformation would you use to uncorrelate the data $\mathbf{X}$? Provide a numerical solution with the information provided. Justify your answer.**

2. (1 point) **Define the linear transformation (with all numerical entries) that Principal Component Analysis (PCA) would use to project this 3-dimensional data to a 2-dimensional representation.**

3. (0.5 points) **What is the amount of explained variance of this 2-D projection? Justify.**

4. (0.5 points) **Let $\mathbf{Y}$ be the data (linear) transformation obtained by principal component transform of $\mathbf{X}$ onto a 2-dimensional space. What is resulting covariance matrix of transformed data $\mathbf{Y}$? Use a numerical answer and justify your answer.**

## Answer:

### 1. Linear transformation to uncorrelate the data (1 point)

To uncorrelate the data $\mathbf{X}$, we need to diagonalize the covariance matrix $\mathbf{K}$. This is achieved by transforming the data using the eigenvector matrix.

**Transformation:** $\mathbf{Z} = \mathbf{V}^T \mathbf{X}$

where $\mathbf{V}$ is the matrix whose columns are the eigenvectors:

$$\mathbf{V} = \begin{bmatrix} -0.99 & 0 & -0.09 \\ 0.09 & 0 & -0.99 \\ 0 & 1 & 0 \end{bmatrix}$$

Therefore, the transformation matrix is:

$$\mathbf{V}^T = \begin{bmatrix} -0.99 & 0.09 & 0 \\ 0 & 0 & 1 \\ -0.09 & -0.99 & 0 \end{bmatrix}$$

**Justification:** The eigenvector matrix $\mathbf{V}$ diagonalizes the covariance matrix (since $\mathbf{K} = \mathbf{V} \mathbf{\Lambda} \mathbf{V}^T$ where $\mathbf{\Lambda}$ is the diagonal matrix of eigenvalues). After transformation, the covariance matrix of $\mathbf{Z}$ becomes $\mathbf{\Lambda}$, which is diagonal, meaning the features are uncorrelated.

### 2. PCA transformation to 2D (1 point)

For PCA, we select the top $k=2$ eigenvectors corresponding to the largest eigenvalues.

**Eigenvalues in descending order:**
- $\lambda_3 = 1.98$ (largest) → eigenvector $v_3 = [-0.09, -0.99, 0]^T$
- $\lambda_1 = 0.98$ (second largest) → eigenvector $v_1 = [-0.99, 0.09, 0]^T$
- $\lambda_2 = 0.5$ (smallest) → eigenvector $v_2 = [0, 0, 1]^T$

The PCA transformation matrix $\mathbf{W}$ is formed by taking the top 2 eigenvectors as columns:

$$\mathbf{W} = \begin{bmatrix} v_3 & v_1 \end{bmatrix} = \begin{bmatrix} -0.09 & -0.99 \\ -0.99 & 0.09 \\ 0 & 0 \end{bmatrix}$$

The linear transformation is: $\mathbf{Y} = \mathbf{W}^T \mathbf{X}$

where:

$$\mathbf{W}^T = \begin{bmatrix} -0.09 & -0.99 & 0 \\ -0.99 & 0.09 & 0 \end{bmatrix}$$

This projects the 3D data $\mathbf{X}$ onto a 2D space $\mathbf{Y}$.

### 3. Explained variance of 2D projection (0.5 points)

**Total variance** = Sum of all eigenvalues = $\lambda_1 + \lambda_2 + \lambda_3 = 0.98 + 0.5 + 1.98 = 3.46$

**Variance retained by top 2 components** = $\lambda_3 + \lambda_1 = 1.98 + 0.98 = 2.96$

**Explained variance ratio** = $\frac{2.96}{3.46} = 0.8555$ or **85.55%**

**Justification:** The explained variance is calculated as the ratio of the sum of the eigenvalues corresponding to the selected principal components to the sum of all eigenvalues. Since we selected the top 2 eigenvectors (with eigenvalues 1.98 and 0.98), we retain 85.55% of the total variance in the original data.

### 4. Covariance matrix of transformed data Y (0.5 points)

When we apply the PCA transformation $\mathbf{Y} = \mathbf{W}^T \mathbf{X}$, the covariance matrix of $\mathbf{Y}$ becomes:

$$\text{Cov}(\mathbf{Y}) = \mathbf{W}^T \mathbf{K} \mathbf{W}$$

Since $\mathbf{W}$ contains the eigenvectors of $\mathbf{K}$, and the eigenvectors are orthogonal, the resulting covariance matrix is diagonal with the corresponding eigenvalues:

$$\text{Cov}(\mathbf{Y}) = \begin{bmatrix} \lambda_3 & 0 \\ 0 & \lambda_1 \end{bmatrix} = \begin{bmatrix} 1.98 & 0 \\ 0 & 0.98 \end{bmatrix}$$

**Justification:** The principal component transformation decorrelates the data, resulting in a diagonal covariance matrix. The diagonal entries are the eigenvalues corresponding to the selected principal components (in order: $\lambda_3$ for the first PC and $\lambda_1$ for the second PC). The off-diagonal elements are zero, indicating that the principal components are uncorrelated.
