# Project C Speech Emotion Classification - Training Dataset

The training dataset contains a total of 648 samples.

```python
import numpy as np
import numpy.random as npr
import random
from IPython.display import display, Audio
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
```

```python
labels_names = ['happy','sad','angry','surprised','neutral']

data_training = np.load('training_data_projectC.npy')
labels_training = np.load('training_labels_projectC.npy')

data_training.shape, labels_training.shape
```

```python
label, counts = np.unique(labels_training, return_counts=True)

plt.bar(label, counts)
plt.title('Sample Size per Class Label')
plt.ylabel('Sample Counts per Label')
plt.xlabel('Labels')
plt.xticks(list(range(1,6)),labels_names,rotation=70);
```

# Loading sample example

```python
# Change the idx for another training sample
idx = 5
plt.figure(figsize=(10,2))
plt.plot(data_training[:,idx])
plt.title('Label: '+labels_names[int(labels_training[idx])-1], fontsize=20);

display(Audio(data_training[:,idx], rate=48000, autoplay=True))
```
