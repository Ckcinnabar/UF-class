## Project E - Simple Object Tracking

The blind test set will be provided as a ````npy```` file with shape of $N\times 15\times 100\times 100\times 3$, where $N$ is sample size.

The extra credit test set will also have a similar shape, $M\times 15\times 100\times 100\times 3$, and it will contain samples from an unknown class (label -1).

In both test sets, the ground truth labels will have a shape of $N\times 15\times 5$. Your model must predict bounding box and classify each object in the each video (15 extracted frames), and calculate the **average intersection-over-union (IoU)**.
* For each video, take the average IoU over all frames **with a correctly classified object label prediction**, average it and weigh it based on number of frames with correct predictions. The final score will be an average of this quantity across entire set.
* For example, video 1 has 15 frames with object label 2. Frames 1-13 were labeled as 2, but frames 14-15 were labeled as 3. Let's say the IoU for frames 1-13 was 0.80 each. Then video 1's average score is 0.80*13/15. Repeat this for every video, and then take the average. This average should be at $\geq0.55$ for full marks on blind test performance.
