**Consider a training dataset $\{(x_i,t_i)\}_{i=1}^N$ where $x_i\in\mathbb{R}$ and $t_i\in\mathbb{R}$.**

**Answer the folowing questions:**

1. (2 points) **Suppose you are training a mapper function of the form:**

$$y(x)=\exp\left(\sum_{j=0}^M w_jx^j\right)$$

**Derive the analytical solution of $\mathbf{w}$ using the Normal Equations and the Mean Squared Error (MSE). Show your work.**

Note that we can find an analytical solution only in the log-space of the target variable, i.e.

\begin{align*}
y(x) &= \exp\left(\sum_{j=0}^M w_jx^j\right) \\
\ln(y(x)) &= \sum_{j=0}^M w_jx^j = \mathbf{X}\mathbf{w}\\
\end{align*}

Now, this is a linear regression problem, and the solution for $\mathbf{w}$ using Normal Equations of the MSE objective function is

$$\mathbf{w} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\ln\left(\mathbf{t}\right)$$

To make predictions, we simply take:

$$y(x) = \exp(\mathbf{X}\mathbf{w})$$

**Consider the train sample (blue circles) and test samples (red traingles) below.**

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

x_train = np.squeeze(np.load('x_train.npy'))
x_test = np.squeeze(np.load('x_test.npy'))
t_train = np.load('t_train.npy')
t_test=np.load('t_test.npy')

plt.plot(x_train, t_train, 'ob', label='Training samples')
plt.plot(x_test, t_test, '^r', label='Test samples')
plt.legend(fontsize=15);
```

3. (2 points) **Use the template code from lecture (copied below), modified it as needed to implement your derived solution for mapper function in part (1).**

```python
import scipy.stats as stats
import numpy as np
```

```python
def exponential_regression(x, t, M):
    # Rename this function and provide the necessary inputs
    
    # Compute feature matrix
    X = np.array([x**i for i in range(M+1)]).T
    
    # Compute the solution for the parameters w
    w = np.linalg.inv(X.T@X)@X.T@np.log(t)
    
    # Compute model prediction
    y = np.exp(X@w)
    
    return w, y
```

```python
#Polynomial order
M = 3 #hard-coding a model order for now
w, y_train = exponential_regression(x_train, t_train, M)

plt.plot(x_train, t_train, 'ob', label='Training samples')
plt.plot(x_train, y_train, 'b', label='Predicted Model')
plt.plot(x_test, t_test, '^r', label='Test samples')
plt.legend(fontsize=15);
```

3. (1 point) **Use your trained models from part (2) to make predictions for the test samples. Discuss your observations.**

```python
def exponential_regression_test(x, w):
    # Rename this function and provide the necessary inputs
    
    # Compute feature matrix
    X = np.array([x**i for i in range(len(w))]).T
    
    # Compute model prediction
    y = np.exp(X@w)
    
    return y
```
